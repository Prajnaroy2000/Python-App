# EXPLORING INSIGHTS THROUGH DATA VISUALIZATION IN THE E-COMMERCE PROJECT

## 1.City-wise Distribution of Customers:

```python
import matplotlib.pyplot as plt
import pandas as pd

customer_city_counts = customer_df['city'].value_counts()
plt.figure(figsize=(10, 6))
customer_city_counts.plot(kind='bar', color='skyblue')
plt.title('City-wise Distribution of Customers')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 2.Order Status Distribution:

```python
import matplotlib.pyplot as plt
import pandas as pd

order_status_counts = order_df['order_status'].value_counts()
plt.figure(figsize=(8, 6))
order_status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['pink', 'lightgreen', 'orange'])
plt.title('Order Status Distribution')
plt.axis('equal')
plt.show()
```

## 3.Most Preferred Payment Modes

```python
import matplotlib.pyplot as plt
import pandas as pd

payment_mode_counts = order_df['payment_mode'].value_counts()
plt.figure(figsize=(10, 6))
payment_mode_counts.plot(kind='bar', color='lightgreen')
plt.title('Most Preferred Payment Modes')
plt.xlabel('Payment Mode')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 4.Product Pricing Distribution

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 6))
product_df[['original_price', 'selling_price']].boxplot()
plt.title('Product Pricing Distribution')
plt.ylabel('Price (in thousands)')
plt.xticks([1, 2], ['Original Price', 'Selling Price'])
plt.tight_layout()
plt.show()
```

## 5.Product Stock Levels

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(10, 6))
product_df.plot(x='product_name', y='stock', kind='bar', color='blue')
plt.title('Product Stock Levels')
plt.xlabel('Product Name')
plt.ylabel('Stock Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 6.Total Revenue by Product Category

```python
import matplotlib.pyplot as plt
import pandas as pd

total_revenue_category = order_df.merge(product_df, on='product_id').groupby('category')['total_price'].sum()
plt.figure(figsize=(12, 6))
plt.bar(total_revenue_category.index, total_revenue_category.values, color='lightcoral')
plt.title('Total Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

## 7.Top 10 Best-selling Products

```python
import matplotlib.pyplot as plt
import pandas as pd

top_10_products = order_df.merge(product_df, on='product_id').groupby('product_name')['quantity'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
plt.barh(top_10_products.index, top_10_products.values, color='gold')
plt.title('Top 10 Best-selling Products')
plt.xlabel('Quantity Sold')
plt.ylabel('Product Name')
plt.show()
```

## 8.Total Revenue by City

```python
import matplotlib.pyplot as plt
import pandas as pd

total_revenue_city = order_df.merge(customer_df, on='customer_id').groupby('city')['total_price'].sum()
plt.figure(figsize=(12, 6))
plt.bar(total_revenue_city.index, total_revenue_city.values, color= 'cyan')
plt.title('Total Revenue by City')
plt.xlabel('City')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

## 9.Total Revenue by Payment Mode

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 8))
payment_mode_revenue = order_df.groupby('payment_mode')['total_price'].sum()
plt.pie(payment_mode_revenue, labels=payment_mode_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Revenue by Payment Mode')
plt.show()
```

## 10.Customer Distribution by Pin Code

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(14, 6))
customer_distribution_pincode = customer_df['pin_code'].value_counts()
plt.bar(customer_distribution_pincode.index.astype(str), customer_distribution_pincode.values, color='lightgreen')
plt.title('Customer Distribution by Pin Code')
plt.xlabel('Pin Code')
plt.ylabel('Number of Customers')
plt.xticks(rotation=90)
plt.show()
```

• The boxplot indicates the spread of original prices across the products in the dataset. The boxplot illustrates the median, interquartile range (IQR), and outliers in the distribution. • By comparing the boxplots of original and selling prices side by side, we can assess how selling prices deviate from the original prices. The comparison helps identify the extent of markups or markdowns applied to the products.
