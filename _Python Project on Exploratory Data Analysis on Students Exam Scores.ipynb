# DATA DICTIONARY

- Gender: Gender of the student (male/female) score(0-100)

- EthnicGroup: Ethnic group of the student (group A to E)

- ParentEduc: Parent(s) education background (from some_highschool to master's degree)


- LunchType: School lunch type (standard or free/reduced)

- TestPrep: Test preparation course followed (completed or none)

- ParentMaritalStatus: Parent(s) marital status (married/single/widowed/divorced)

- PracticeSport: How often the student parctice sport (never/sometimes/regularly))

- IsFirstChild: If the child is first child in the family or not (yes/no)

- NrSiblings: Number of siblings the student has (0 to 7)

- TransportMeans: Means of transport to school (schoolbus/private)

- WklyStudyHours: Weekly self-study hours(less that 5hrs; between 5 and 10hrs; more than 10hrs)

- MathScore: math test score(0-100)

- ReadingScore: reading test score(0-100)

- WritingScore: writing test score(0-100)

- ParentMaritalStatus: Parent(s) marital status (married/single/widowed/divorced)

# EXPLORATORY DATA ANANLYSIS:

### COUNTING THE NUMBER OF STUDENTS BY GENDER:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "new_cleaned_Expanded_data_with_more_features.csv.zip"

df = pd.read_csv(file_path)

# Calculate the gender counts
gender_counts = df['Gender'].value_counts()

# Plot the gender distribution using a bar plot
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=gender_counts.index, y=gender_counts, hue=gender_counts.index, palette='viridis', legend=False )  
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### INFERENCE

-The plot displays the distribution of genders (presumably 'Male' and 'Female') in your dataset.

-We can observe the count of each gender category represented by the bars on the plot.


### ANALYSING THE AVERAGE SCORE OF EACH STUDENT: 

```python
import pandas as pd

file_path = 'Expanded_data_with_more_features.csv.zip'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
df = df.drop("Unnamed: 0", axis = 1)
df.head()
# Calculate the average score across the three columns
df['Average_Score'] = df[['MathScore', 'ReadingScore', 'WritingScore']].mean(axis=1)
df
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Average_Score', data=df, palette='muted')
plt.title('Average Score of Each Student')
plt.xlabel('Gender')
plt.ylabel('Average_Score')
plt.show()
```

#### INFERENCE 

-The bar plot provides a visual comparison of the average scores between male and female students.
-Weu can observe whether there are any significant differences in average scores based on gender
.-We can concluded that average marks of female is grater than male.n.

### ANALYSIS THE NUMBER OF STUDENTS TAKE PREPARATORY CLASSES:

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Expanded_data_with_more_features.csv.zip'
df = pd.read_csv(file_path, compression='zip')
count_test = df['TestPrep'].value_counts()
labels = df['TestPrep'].value_counts().index
colors = ['skyblue', 'lightcoral']
plt.figure(figsize= (6,6))
plt.pie(count_test, labels = labels, autopct='%1.2f%%', colors=colors)
plt.legend(labels)
plt.show()
```

### ANALYSING HOW THE EDUCATION OF PARENTS IMPACT ON STUDENTS SCORE 

```python
import pandas as pd
file_path = 'Expanded_data_with_more_features.csv.zip'
df = pd.read_csv(file_path)
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
file_path = 'Expanded_data_with_more_features.csv.zip'
df = pd.read_csv(file_path)
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show
```

#### INFERENCE:

- From the above charte we have concluded that the education of the parents have a good impact on their scores

### DISTRIBUTION OF ETHNIC GROUPS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'Expanded_data_with_more_features.csv.zip'
df = pd.read_csv(file_path)
df = df.drop("Unnamed: 0", axis = 1)
GroupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
GroupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
GroupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
GroupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
GroupE = df.loc[(df['EthnicGroup'] == 'group E')].count()
mylist=[GroupA['EthnicGroup'], GroupB['EthnicGroup'], GroupC['EthnicGroup'], GroupD['EthnicGroup'], GroupE['EthnicGroup']]
colors = ['blue', 'orange', 'green', 'red', 'purple']
labels = ['group C', 'group B', 'group A', 'group D', 'group E']
plt.pie(mylist,labels=labels, colors=colors, autopct='%1.2f%%', startangle=90)
plt.axis('equal')
plt.show()
```

#### INFERENCE

- Maximum ethnic group is group, 31.98% and minimum value is in ethnic group c, 7.70%

### EXPLORE THE RELATIONSHIP BETWEEN STUDY HOURS AND MATH SCORE

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_path = 'Expanded_data_with_more_features.csv.zip'
df = pd.read_csv(file_path)

# Scatter plot to explore the relationship between Study Hours and Scores
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WklyStudyHours',hue='Gender', y='MathScore', data=df)
plt.title('Relationship Between Study Hours and Scores')
plt.xlabel('WklyStudyHours')
plt.ylabel('MathScore')
plt.legend(title='Gender')
plt.show()
```

### ANALYSIS THAT IS THERE ANY IMPACT ON STUDENT SCORE BY TRANSPORT MEANS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = 'Expanded_data_with_more_features.csv.zip'

df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
palette = {'school_bus': 'yellow', 'private':'green'}
sns.barplot(x='TransportMeans', y='ReadingScore', palette=palette, data=df, ci=None)


plt.title('Impact of Transportation on Reading Scores')
plt.xlabel('TransportMeans')
plt.ylabel('ReadingScore')

plt.show()
```

#### INFERENCE

- The bar plot shows the average reading scores for different transportation means, represented by the 'school_bus' and 'private'.

- 
By comparing the heights of the bars, you can infer the average reading scores associated with each transportation category

- Here we can see there is no impact of transport means on students reading scores..
